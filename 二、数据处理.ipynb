## 1.4.获取院校经纬度信息

```python
import pandas as pd
import requests
import time
def get_geocode(address, ak):
    url = 'http://api.map.baidu.com/geocoding/v3/'
    params = {
        'address': address,
        'output': 'json',
        'ak': ak
    }
    response = requests.get(url, params=params)
    data = response.json()
    if data['status'] == 0:
        location = data['result']['location']
        return location['lng'], location['lat']
    else:
        return None
    time.sleep(0.5)

ak = '2XXwGmMnh2yoi8GHhQZrhVzLxENnetyB'

# 添加经纬度列
data['经度'] = None
data['纬度'] = None

# 遍历数据集中的每行院校名称
for index, row in data.iterrows():
    college = row['院校名称']
    try:
        lng, lat = get_geocode(college, ak)
        if lng and lat:
            data.at[index, '经度'] = lng
            data.at[index, '纬度'] = lat
            print(college, ':', lng, lat)
            
    except:
        data.at[index, '经度'] = '未知'
        data.at[index, '纬度'] = '未知'
        print(college, ':', "未知", "未知")

print('已完成')
```

```python
data.head()
```

```python
data.to_csv('国内大学_列处理.csv',index=0)
```

# 2.数据清洗

## 2.1.对类别特征进行硬编码

```python
data=pd.read_csv('国内大学_列处理.csv')
```

```python
data=data.fillna(-1)
```

```python
data.columns
```

```python
data.replace('未知', 0, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
def get_feat1(df, lst):
    for feat in lst:
        enc = LabelEncoder()
        df[feat] = df[feat].astype('str')
        df[feat] = enc.fit_transform(df[feat])
    return df
feat_lst = ['是否是211', '是否是985', '是否有研究生院', '院校隶属', '所在地区','办学层次','学历','类型']
data = get_feat1(data, feat_lst)
```

```python
data.to_csv('国内大学_硬编码.csv',index=0)
```

## 2.2.标准化数据

```python
data=pd.read_csv('国内大学_硬编码.csv')
```

```python
def min_max_scaler(x):
    return (x-x.min())/(x.max()-x.min())

def stander_scaler(x):
    return (x-x.mean())/x.std()

def decimal_scaler(x):
    import numpy as np
    return x/10**(np.ceil(np.log10(x.abs().max())))
```

```python
data1=decimal_scaler(data.iloc[:, 1:])
```

```python
data2=pd.concat([data.iloc[:,0], data1], axis=1)
```
